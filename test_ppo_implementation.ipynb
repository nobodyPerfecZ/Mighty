Then let's load the data:

```python
import pandas as pd

def load_results(experiment_name, seeds=[0,1,2,3,4]):
    dfs = []
    eval_dfs = []
    for s in seeds:
        eval_df = pd.read_csv(f"runs/{experiment_name}_{s}/eval_results.csv")
        df = pd.read_csv(f"runs/{experiment_name}_{s}/results.csv")
        dfs.append(df)
        eval_dfs.append(eval_df)
    return pd.concat(dfs), pd.concat(eval_dfs)
```

```python
ppo_cartpole, eval_cartpole = load_results("cartpole_ppo_latest", seeds=[0])

import seaborn as sns
sns.lineplot(x=eval_cartpole["step"], y=eval_cartpole["mean_eval_reward"])
```
